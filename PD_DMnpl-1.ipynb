### drop_na(it's parameter)

```python
df1=pd.read_excel(r'D:\Practice Pandas\Test1.xlsx', sheet_name='Sheet6')
df1
```

```python
df1.dropna()         # bydefault it will remove all the rows which contain atleast one nan values
```

```python
df1.dropna(axis=1)         # bydefault it will remove all the columns which contain atleast one nan values
```

```python
df1.dropna(how='all')     # bydefault it will remove the rows which contain all the values as null.
```

```python
df1.dropna(thresh=4)            # bydefault it will keep the rows which contain minimum 4 non-NaN values.
```

```python
df1.dropna(subset={'Class'})              #subset : array-like, optional Labels along other axis to consider, 
                                         # e.g. if you are dropping rows these would be a list of columns to include.
```

```python
df1.dropna(subset={'Roll Number'})
```

### replace(['to_replace=None', 'value=None', 'inplace=False', 'limit=None', 'regex=False', "method='pad'"],) function and its parameter

```python
df=pd.read_excel(r'D:\Practice Pandas\Test1.xlsx', sheet_name='Sheet2')
df
```

```python
df.replace(to_replace='ten', value='one')
```

```python
df.replace('ten','one')         # can be written like this also
```

```python
df.replace(25,2500)    # replacing nueric value
```

```python
df.replace([25,26,27,28,29,30,31,32,33], 50)     #replacing total value in ID column with a single value 50
```

```python
df.replace(['aj', 'bj', 'cj', 'dj'], 'shailesh') # same way we can replace the string value too
```

```python
df.replace([25,26,27,28,29,30,31,32,33], [55,56,57,58,59,60,61,62,63]) #replacing total value in ID column with different values
```

```python
df.replace({'Class':'seven'}, 'nine') # also by using the dictionay we can select a particular column's value and replace that
```

```python
df.replace({'Class':'seven' , 'Roll Number': 14}, 'nine')  # same way by using the dictionay we can select multiple column's value and replace with same value
```

### regex

####   we can convert all the charactor(A-z and a-z) into some specific number or any other charactor using regex

```python
df.replace('[A-Za-z]', 0 , regex=True)   # all charactor converted into 0 value
```

```python
df.replace('[A-Za-z]', 'W' , regex=True)
```

```python
df.replace({'Name': '[A-Za-z]'}, 'W' , regex=True)  # same way we can select the particular columns too
```
